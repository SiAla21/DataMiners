**Load and Inspect Data:**

```python
import pandas as pd
import numpy as np
import unicodedata
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
```

```python
Doctor_df = pd.read_csv('/content/Doctor.csv')
```

```python
# Check the first few rows of the datasets
Doctor_df.head()
```

```python
print("General data information :")
display(Doctor_df.info())
```

```python
display(Doctor_df.describe())
```

```python
nan_counts = Doctor_df.isna().sum().sum()
print(nan_counts)
```

```python
print("\nMissing values ​​before processing :")
display(Doctor_df.isnull().sum())
```

```python
# ➤ Automatic processing: we fill the numeric columns with the average
num_cols = Doctor_df.select_dtypes(include=[np.number]).columns
for col in num_cols:
    Doctor_df[col].fillna(Doctor_df[col].mean(), inplace=True)
```

```python
#For text columns, you can fill with 'unknown' if necessary
text_cols = Doctor_df.select_dtypes(include=['object']).columns
for col in text_cols:
    Doctor_df[col].fillna('inconnu', inplace=True)
```

```python
# Outliers on numeric columns
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    before = df.shape[0]
    df_clean = df[(df[column] >= lower) & (df[column] <= upper)]
    print(f"✅ {column} : {before - df_clean.shape[0]} outliers removed")
    return df_clean
```

```python
# Apply only to relevant numeric columns
for col in num_cols:
    med_df = remove_outliers_iqr(Doctor_df, col)
```

```python
#Removing duplicates
initial_shape = Doctor_df.shape
Doctor_df.drop_duplicates(inplace=True)
print(f"🔁 Duplicates removed : {initial_shape[0] - med_df.shape[0]} lines deleted")
```

```python
# ✅ Finalisation
print("\nDoctor data ready after pre-processing:")
display(Doctor_df.head())
```

**Data Transformation:**

```python
# Replace negative values ​​with their absolute values
for col in num_cols:
    if (Doctor_df[col] < 0).any():
        print(f"Correction of negative values ​​in column: {col}")
        Doctor_df[col] = Doctor_df[col].abs()
```

```python
# Checking that there are no more negative values
print("\nChecking for remaining negative values :")
print((Doctor_df[num_cols] < 0).sum())
```

```python
# Select columns with object or category data types (categorical columns)
categorical_columns = Doctor_df.select_dtypes(include=['object', 'category'])

print(Doctor_df.columns.tolist())

# Display the names of categorical columns
print(categorical_columns.columns.tolist())
```

```python
# List of provided categorical columns
categorical_columns = [
    'Rndrng_Prvdr_Last_Org_Name', 'Rndrng_Prvdr_First_Name', 'Rndrng_Prvdr_MI',
    'Rndrng_Prvdr_Crdntls', 'Rndrng_Prvdr_Ent_Cd', 'Rndrng_Prvdr_St1',
    'Rndrng_Prvdr_St2', 'Rndrng_Prvdr_City', 'Rndrng_Prvdr_State_Abrvtn',
    'Rndrng_Prvdr_State_FIPS', 'Rndrng_Prvdr_Zip5', 'Rndrng_Prvdr_RUCA_Desc',
    'Rndrng_Prvdr_Cntry', 'Rndrng_Prvdr_Type', 'Rndrng_Prvdr_Mdcr_Prtcptg_Ind',
    'Drug_Sprsn_Ind', 'Med_Sprsn_Ind'
]

# Apply LabelEncoder
label_encoders = {}
for col in categorical_columns:
    le = LabelEncoder()
    Doctor_df[col] = le.fit_transform(Doctor_df[col].astype(str))  # Cast to str in case of NaNs
    label_encoders[col] = le
```

```python
print("Preview after transformation of categorical columns :")
display(Doctor_df.head())

print("\nColumn types after encoding :")
print(Doctor_df.dtypes)
```

**Data normalization:**

```python
# We select the numeric columns (excluding Booleans if there are any)
numeric_cols = Doctor_df.select_dtypes(include=[np.number]).columns.tolist()
print(f"Numerical columns to be normalized : {numeric_cols}")
```

```python
# Encoded categorical columns that are excluded from normalization
excluded_encoded_cols = [
    'Rndrng_Prvdr_Last_Org_Name', 'Rndrng_Prvdr_First_Name', 'Rndrng_Prvdr_MI',
    'Rndrng_Prvdr_Crdntls', 'Rndrng_Prvdr_Ent_Cd', 'Rndrng_Prvdr_St1', 'Rndrng_Prvdr_St2',
    'Rndrng_Prvdr_City', 'Rndrng_Prvdr_State_Abrvtn', 'Rndrng_Prvdr_State_FIPS',
    'Rndrng_Prvdr_Zip5', 'Rndrng_Prvdr_RUCA_Desc', 'Rndrng_Prvdr_Cntry',
    'Rndrng_Prvdr_Type', 'Rndrng_Prvdr_Mdcr_Prtcptg_Ind', 'Drug_Sprsn_Ind', 'Med_Sprsn_Ind'
]

# Truly continuous numeric columns to be normalized
cols_to_normalize = [col for col in numeric_cols if col not in excluded_encoded_cols]
```

```python
scaler = MinMaxScaler()
Doctor_df[cols_to_normalize] = scaler.fit_transform(Doctor_df[cols_to_normalize])
```

```python
print("Example of normalized data :")
display(Doctor_df[cols_to_normalize].head())

print("Post-normalization statistics :")
display(Doctor_df[cols_to_normalize].describe())
```
