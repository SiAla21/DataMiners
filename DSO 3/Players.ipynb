```python
def detect_outliers_iqr(df, columns):
    outliers = {}
    for col in columns:
        Q1 = df[col].quantile(0.25)  # Premier quartile (25%)
        Q3 = df[col].quantile(0.75)  # Troisième quartile (75%)
        IQR = Q3 - Q1  # Intervalle interquartile
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]
    return outliers

# Select the numeric columns to analyze
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Outlier detection
outliers_dict = detect_outliers_iqr(df, numeric_cols)

# Show columns that have outliers
for col, outliers in outliers_dict.items():
    if not outliers.empty:
        print(f"⚠ Outliers detected in {col}:")
        print(outliers)
```

#**II.Data Preparation 🛠️**

#1. Data Cleaning 🧹:

1️⃣ **Handle missing values :**

**Numerical columns:** Replace with the median to avoid bias.

**Categorical columns:** Replace with the most frequent modality.

```python
# Replacing missing values ​​in numeric columns with the median
df.fillna(df.median(numeric_only=True), inplace=True)

# Replacing missing values ​​in categorical columns with the most frequent value
for col in df.select_dtypes(include=['object']).columns:
    df[col].fillna(df[col].mode()[0], inplace=True)

print("Missing values ​​after cleaning :")

# Verification
print(df.isnull().sum())
```

2️⃣ **Remove duplicates:**

```python
df.drop_duplicates(inplace=True)
print(f"Number of duplicates after cleaning : {df.duplicated().sum()}")
```

 3️⃣ **Replace_outliers_with_mode:**

```python
def replace_outliers_with_mode(df, columns):
    """
    Detects outliers in each column using the IQR method and replaces them with the most frequent value.
    """
    for col in columns:
        # Calculate Q1, Q3, and IQR
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1

        # Define lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Identify the most frequent value (mode)
        mode_value = df[col].mode()[0]

        # Replace outliers with mode
        df.loc[(df[col] < lower_bound) | (df[col] > upper_bound), col] = mode_value

    return df

# Select numeric columns
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Replace outliers with mode
df = replace_outliers_with_mode(df, numeric_cols)

print("✅ Outliers have been replaced with the most frequent value in each column.")
```

*   Correlation Matrix:




```python
# Select numeric columns
X = df.select_dtypes(include=['number'])

# Compute correlation matrix
corr_matrix = X.corr().abs()

# Set figure size and style
plt.figure(figsize=(14, 10))
sns.set_style("white")

# Create the heatmap with optimized settings
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=1,
            annot_kws={"size": 9}, cbar_kws={"shrink": 0.8}, center=0, square=True)

# Rotate labels for better readability
plt.xticks(rotation=45, ha="right", fontsize=10, color='black')
plt.yticks(fontsize=10, color='black')

# Adjust the title with better styling
plt.title("🔥 Improved Correlation Matrix", fontsize=16, fontweight='bold', color='darkblue')

# Display the heatmap
plt.show()
```

* Highly correlated column

```python
# Define the correlation threshold (e.g., 0.8 for strong correlation)
threshold = 0.8

# Find pairs of highly correlated columns
high_corr_pairs = []
for i in range(len(corr_matrix.columns)):
    for j in range(i + 1, len(corr_matrix.columns)):  # Avoid duplicate pairs
        if corr_matrix.iloc[i, j] > threshold:
            high_corr_pairs.append([corr_matrix.index[i], corr_matrix.columns[j], round(corr_matrix.iloc[i, j], 2)])

# Display results in a structured table
if high_corr_pairs:
    print("\n📊 Highly Correlated Columns (Threshold > 0.8):\n")
    print(tabulate(high_corr_pairs, headers=["Column 1", "Column 2", "Correlation"], tablefmt="psql"))
else:
    print("\n✅ No highly correlated columns found above the threshold.")
```
