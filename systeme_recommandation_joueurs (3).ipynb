# ðŸ“Š SystÃ¨me de recommandation de joueurs pour clubs

```python

import pandas as pd
##
# Charger les bases de donnÃ©es (Ã  adapter au chemin si nÃ©cessaire)
players_df = pd.read_csv('Players After Cleaning.csv', encoding='utf-8')
clubs_df = pd.read_csv('all_leagues.csv', encoding='utf-8')

players_df.head(), clubs_df.head()
```

### âž¡ï¸ Fonction de scoring pondÃ©rÃ© selon le poste

```python

def score_joueur(joueur, poste):
    score = joueur['overall_rating'] * 0.3
    age_penalty = max(0, (joueur['age'] - 28) * 0.5)  # pÃ©nalitÃ© pour les joueurs de plus de 28 ans

    if poste in ['CB', 'LB', 'RB']:
        score += joueur.get('defensive_awareness', 0) * 0.25
        score += joueur.get('interceptions', 0) * 0.2
        score += joueur.get('tackling', 0) * 0.2
    elif poste in ['ST', 'CF']:
        score += joueur.get('finishing', 0) * 0.25
        score += joueur.get('shot_power', 0) * 0.2
        score += joueur.get('positioning', 0) * 0.2
    elif poste in ['CM', 'CAM']:
        score += joueur.get('passing', 0) * 0.3
        score += joueur.get('vision', 0) * 0.25
        score += joueur.get('ball_control', 0) * 0.2
    elif poste == 'GK':
        score += joueur.get('reflexes', 0) * 0.3
        score += joueur.get('handling', 0) * 0.25
        score += joueur.get('positioning', 0) * 0.2

    score -= age_penalty
    return score
```

### âž¡ï¸ Fonction de recommandation

```python
def recommander_joueurs(club_name, top_n=5):
    """Recommande les meilleurs joueurs Ã  recruter pour un club en fonction de ses faiblesses."""
    
    # VÃ©rifier si le club existe dans clubs_df
    if club_name not in clubs_df['club'].values:
        print(f"âŒ Club '{club_name}' introuvable dans clubs_df.")
        return {}

    club = clubs_df[clubs_df['club'] == club_name].iloc[0]
    recommandations = {}

    # VÃ©rifier les valeurs du club
    print("\nðŸ”¹ VÃ©rification des statistiques du club :")
    print(f"âž¡ Goals Conceded : {club['goals_conceded']}")
    print(f"âž¡ Goals Scored : {club['goals_scored']}")
    print(f"âž¡ Possession (%) : {club['possession(%)']}")
    print(f"âž¡ Clean Sheets : {club['clean_sheets']}")

    # VÃ©rifier si players_df contient bien la colonne 'registered_position'
    if 'registered_position' not in players_df.columns:
        print("âŒ Erreur : La colonne 'registered_position' est absente de players_df.")
        return {}

    # VÃ©rification des postes disponibles
    print("\nðŸ”¹ Postes disponibles dans players_df :", players_df['registered_position'].unique())

    # DÃ©fense faible
    if club['goals_conceded'] > 40:
        defenseurs = players_df[players_df['registered_position'].isin(['CB', 'LB', 'RB'])].copy()
        if not defenseurs.empty:
            defenseurs['score'] = defenseurs.apply(lambda x: score_joueur(x, x['registered_position']), axis=1)
            defenseurs = defenseurs.sort_values(by='score', ascending=False).head(top_n)
            recommandations['DÃ©fenseurs'] = defenseurs[['name', 'registered_position', 'defensive_awareness', 'interceptions', 'tackling', 'age', 'overall_rating', 'score']]
    
    # Attaque faible
    if club['goals_scored'] < 30:
        attaquants = players_df[players_df['registered_position'].isin(['ST', 'CF'])].copy()
        if not attaquants.empty:
            attaquants['score'] = attaquants.apply(lambda x: score_joueur(x, x['registered_position']), axis=1)
            attaquants = attaquants.sort_values(by='score', ascending=False).head(top_n)
            recommandations['Attaquants'] = attaquants[['name', 'registered_position', 'finishing', 'shot_power', 'positioning', 'age', 'overall_rating', 'score']]
    
    # Milieu faible
    if club['possession(%)'] < 50:
        milieux = players_df[players_df['registered_position'].isin(['CM', 'CAM'])].copy()
        if not milieux.empty:
            milieux['score'] = milieux.apply(lambda x: score_joueur(x, x['registered_position']), axis=1)
            milieux = milieux.sort_values(by='score', ascending=False).head(top_n)
            recommandations['Milieux crÃ©atifs'] = milieux[['name', 'registered_position', 'passing', 'vision', 'ball_control', 'age', 'overall_rating', 'score']]
    
    # Gardien si besoin
    if club['clean_sheets'] < 5:
        gardiens = players_df[players_df['registered_position'] == 'GK'].copy()
        if not gardiens.empty:
            gardiens['score'] = gardiens.apply(lambda x: score_joueur(x, x['registered_position']), axis=1)
            gardiens = gardiens.sort_values(by='score', ascending=False).head(top_n)
            recommandations['Gardiens'] = gardiens[['name', 'reflexes', 'handling', 'positioning', 'age', 'overall_rating', 'score']]
    
    return recommandations
```

### âœ… Exemple d'utilisation

```python
# SÃ©lection d'un club pour le test
club_exemple = clubs_df.iloc[0]['club']
print(f"\nâœ… SÃ©lection du club : {club_exemple}")

# Obtenir les recommandations
recommandations = recommander_joueurs(club_exemple)

# Affichage des rÃ©sultats
if not recommandations:
    print("\nâŒ Aucune recommandation trouvÃ©e pour ce club.")
else:
    for categorie, joueurs in recommandations.items():
        print(f"\nðŸ”¹ Recommandations pour {categorie} :")
        print(joueurs)
```

```python
import pandas as pd

# Charger les bases de donnÃ©es
players_df = pd.read_csv('Players After Cleaning.csv', encoding='utf-8')
clubs_df = pd.read_csv('all_leagues.csv', encoding='utf-8')

# Fonction pour calculer le score selon poste et stats club
def score_joueur(joueur, poste, possession=50.0):
    score = joueur['overall_rating'] * 0.3
    age_penalty = max(0, (joueur['age'] - 28) * 0.5)

    if poste in ['CB', 'LB', 'RB']:
        score += joueur.get('ball_winning', 0) * 0.25
        score += joueur.get('defensive_awareness', 0) * 0.2
        score += joueur.get('aggression', 0) * 0.2

    elif poste in ['ST', 'CF']:
        score += joueur.get('finishing', 0) * 0.25
        score += joueur.get('kicking_power', 0) * 0.2
        score += joueur.get('offensive_awareness', 0) * 0.2

    elif poste in ['CMF', 'AMF']:
        pass_weight = 0.3 + (52 - possession) * 0.01
        pass_weight = max(0.3, min(pass_weight, 0.5))
        score += joueur.get('low_pass', 0) * pass_weight
        score += joueur.get('dribbling', 0) * 0.25
        score += joueur.get('ball_control', 0) * 0.2

    elif poste == 'GK':
        score += joueur.get('gk_reflexes', 0) * 0.3
        score += joueur.get('gk_catching', 0) * 0.25
        score += joueur.get('gk_awareness', 0) * 0.2

    score -= age_penalty
    return score

# Recommandation pour un club
def recommander_joueurs(club_name, top_n=5):
    club = clubs_df[clubs_df['club'] == club_name].iloc[0]
    print(f"\nSÃ©lection du club : {club_name}")
    print(f"ðŸ”¹ Statistiques : GC: {club['goals_conceded']} | GS: {club['goals_scored']} | Poss: {club['possession(%)']} | Clean Sheets: {club['clean_sheets']}")
    print(f"ðŸ“Œ Postes trouvÃ©s : {players_df['registered_position'].unique()}")

    recommandations = {}

    # DÃ©fenseurs
    if club['goals_conceded'] > 45:
        defenseurs = players_df[players_df['registered_position'].isin(['CB', 'LB', 'RB'])].copy()
        defenseurs['score'] = defenseurs.apply(lambda x: score_joueur(x, x['registered_position']), axis=1)
        defenseurs = defenseurs.sort_values(by='score', ascending=False).head(top_n)
        recommandations['DÃ©fenseurs'] = defenseurs[['name', 'registered_position', 'defensive_awareness', 'ball_winning', 'aggression', 'age', 'overall_rating', 'score']]

    # Attaquants
    if club['goals_scored'] < 50:
        attaquants = players_df[players_df['registered_position'].isin(['ST', 'CF'])].copy()
        attaquants['score'] = attaquants.apply(lambda x: score_joueur(x, x['registered_position']), axis=1)
        attaquants = attaquants.sort_values(by='score', ascending=False).head(top_n)
        recommandations['Attaquants'] = attaquants[['name', 'registered_position', 'finishing', 'kicking_power', 'offensive_awareness', 'age', 'overall_rating', 'score']]

    # Milieux
    if club['possession(%)'] < 52:
        milieux = players_df[players_df['registered_position'].isin(['CMF', 'AMF'])].copy()
        print(f"ðŸŽ¯ Milieux analysÃ©s : {milieux.shape[0]}")
        milieux['score'] = milieux.apply(lambda x: score_joueur(x, x['registered_position'], club['possession(%)']), axis=1)
        milieux = milieux.sort_values(by='score', ascending=False).head(top_n)
        recommandations['Milieux crÃ©atifs'] = milieux[['name', 'registered_position', 'low_pass', 'dribbling', 'ball_control', 'age', 'overall_rating', 'score']]

    # Gardiens
    if club['clean_sheets'] < 10:
        gardiens = players_df[players_df['registered_position'] == 'GK'].copy()
        gardiens['score'] = gardiens.apply(lambda x: score_joueur(x, x['registered_position']), axis=1)
        gardiens = gardiens.sort_values(by='score', ascending=False).head(top_n)
        recommandations['Gardiens'] = gardiens[['name', 'gk_reflexes', 'gk_catching', 'gk_awareness', 'age', 'overall_rating', 'score']]

    # Affichage
    if not recommandations:
        print("âŒ Aucune recommandation")
    else:
        for cat, joueurs in recommandations.items():
            print(f"\nðŸ”½ Recommandations pour {cat} :")
            print(joueurs)

# Recommandations pour tous les clubs
for club_name in clubs_df['club'].unique():
    print("=" * 70)
    recommander_joueurs(club_name)
    print("=" * 70)
```
