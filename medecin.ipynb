```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
file_path = "D:\Data_tacticsense\doctors_list.csv"
df = pd.read_csv(file_path)

# Display basic info
display(df.info())
display(df.head())
```

```python
# Check for missing values
missing_values = df.isnull().sum()
display(missing_values)

# Visualizing missing values
plt.figure(figsize=(8, 4))
sns.heatmap(df.isnull(), cbar=False, cmap='viridis')
plt.title("Missing Values in Dataset")
plt.show()
```

```python
# Distribution of specialties
plt.figure(figsize=(10, 5))
df['Specialty'].value_counts().plot(kind='bar', color='skyblue')
plt.title("Distribution of Specialties")
plt.xlabel("Specialty")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# Distribution of locations
plt.figure(figsize=(7, 2))
df['Location'].value_counts().head(20).plot(kind='bar', color='salmon')
plt.title("Top 20 Locations")
plt.xlabel("Location")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.show()
```

```python
# Exploring correlations (if numerical features exist)
if df.select_dtypes(include=['number']).shape[1] > 0:
    plt.figure(figsize=(8, 6))
    sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt=".2f")
    plt.title("Correlation Matrix")
    plt.show()
else:
    print("No numerical columns found for correlation analysis.")
```
