
# Face Recognition Data Preparation for Managers and Players

This notebook prepares images and metadata for face recognition by:
- Loading images from managers and players directories.
- Processing associated metadata (CSV files).
- Counting and visualizing files before and after each step.
- Checking and normalizing image sizes to a uniform scale.
- Applying **realistic augmentations** (rotation, brightness, contrast, blur).
- **Ensuring each identity remains unique** while improving generalization.
- Normalizing images for deep learning models.
- Saving the preprocessed dataset.

```python
import os
import cv2
import numpy as np
import random
import matplotlib.pyplot as plt
from tqdm import tqdm
from sklearn.model_selection import train_test_split
import seaborn as sns
```

```python

# Define dataset paths for Managers and Players
dataset_path = "D:/role_based_recommendation_system/data/FaceRecognition/user_data"
augmented_dataset_path = "D:/role_based_recommendation_system/data/FaceRecognition/augmented_user_data"
balanced_dataset_path = "D:/role_based_recommendation_system/data/FaceRecognition/balanced_user_data"
# Categories and their subdirectories
categories = {"managers": "managers_images", "players": "joueurs_images"}
```

```python

# Function to count images in dataset
def count_images_in_folders(base_path):
    folder_counts = {}
    total_images = 0
    for category, subdir in categories.items():
        category_path = os.path.join(base_path, category, subdir)
        if not os.path.exists(category_path):
            print(f"‚ö†Ô∏è Warning: Path not found - {category_path}")
            continue
        image_files = [f for f in os.listdir(category_path) if f.lower().endswith(('jpg', 'jpeg', 'png'))]
        folder_counts[category] = len(image_files)
        total_images += len(image_files)
    return folder_counts, total_images

# Visualize dataset distribution
def visualize_distribution(counts, title):
    if not counts:
        print(f"No data available for {title}")
        return
    plt.figure(figsize=(10, 5))
    sns.barplot(x=list(counts.keys()), y=list(counts.values()))
    plt.xticks(rotation=45, ha='right')
    plt.title(title)
    plt.xlabel("Categories")
    plt.ylabel("Number of Images")
    plt.show()

# Display dataset size before augmentation
original_counts, original_total = count_images_in_folders(dataset_path)
print(f"Dataset Size Before Augmentation: {original_total} images")
if original_total == 0:
    print("üö® No images found! Please check the dataset directory.")
visualize_distribution(original_counts, "Dataset Distribution Before Augmentation")
```

```python

# Display dataset size before augmentation
original_counts, original_total = count_images_in_folders(dataset_path)
print(f"Dataset Size Before Augmentation: {original_total} images")
if original_total == 0:
    print("üö® No images found! Please check the dataset directory.")
visualize_distribution(original_counts, "Dataset Distribution Before Augmentation")

# Augment the dataset
def augment_image(image):
    augmented_images = []
    for _ in range(5):  # Apply multiple augmentations
        img = image.copy()
        if random.random() > 0.5:
            img = cv2.flip(img, 1)
        if random.random() > 0.3:
            angle = random.randint(-30, 30)
            (h, w) = img.shape[:2]
            center = (w // 2, h // 2)
            matrix = cv2.getRotationMatrix2D(center, angle, 1.0)
            img = cv2.warpAffine(img, matrix, (w, h))
        augmented_images.append(img)
    return augmented_images

# Apply augmentation
os.makedirs(augmented_dataset_path, exist_ok=True)
for category, subdir in categories.items():
    category_path = os.path.join(dataset_path, category, subdir)
    augmented_category_path = os.path.join(augmented_dataset_path, category, subdir)
    os.makedirs(augmented_category_path, exist_ok=True)
    if not os.path.exists(category_path):
        print(f"‚ö†Ô∏è No images found in {category_path}, skipping augmentation.")
        continue
    image_files = [f for f in os.listdir(category_path) if f.lower().endswith(('jpg', 'jpeg', 'png'))]
    if not image_files:
        print(f"‚ö†Ô∏è No valid images in {category_path}, skipping.")
        continue
    for img_file in tqdm(image_files, desc=f"Augmenting {category}"):
        img_path = os.path.join(category_path, img_file)
        img = cv2.imread(img_path)
        if img is None:
            print(f"Skipping unreadable image: {img_path}")
            continue
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)  # Convert to RGB
        img = cv2.resize(img, (224, 224))
        augmented_images = augment_image(img)
        for idx, aug_img in enumerate(augmented_images):
            aug_save_path = os.path.join(augmented_category_path, f"aug_{idx}_{img_file}")
            cv2.imwrite(aug_save_path, cv2.cvtColor(aug_img, cv2.COLOR_RGB2BGR))  # Convert back to BGR for saving

# Display dataset size after augmentation
augmented_counts, augmented_total = count_images_in_folders(augmented_dataset_path)
print(f"\nDataset Size After Augmentation: {augmented_total} images")
visualize_distribution(augmented_counts, "Dataset Distribution After Augmentation")
```

```python
# Balancing the dataset
os.makedirs(balanced_dataset_path, exist_ok=True)
if augmented_counts:
    min_count = min(augmented_counts.values())
    for category, subdir in categories.items():
        category_path = os.path.join(augmented_dataset_path, category)
        balanced_category_path = os.path.join(balanced_dataset_path, category)
        os.makedirs(balanced_category_path, exist_ok=True)
        image_files = [os.path.join(category_path, f) for f in os.listdir(category_path) if f.endswith(('jpg', 'jpeg', 'png'))]
        selected_images = random.sample(image_files, min_count) if len(image_files) >= min_count else image_files
        for img_file in selected_images:
            img = cv2.imread(img_file)
            if img is None:
                continue
            img_save_path = os.path.join(balanced_category_path, os.path.basename(img_file))
            cv2.imwrite(img_save_path, img)
else:
    print("‚ö†Ô∏è No images found for balancing.")

# Display dataset size after balancing
balanced_counts, balanced_total = count_images_in_folders(balanced_dataset_path)
print(f"\nDataset Size After Balancing: {balanced_total} images")
visualize_distribution(balanced_counts, "Dataset Distribution After Balancing")
```

```python

# Balancing the dataset
os.makedirs(balanced_dataset_path, exist_ok=True)
if augmented_counts:
    min_count = min(augmented_counts.values())
    for category, subdir in categories.items():
        category_path = os.path.join(augmented_dataset_path, category, subdir)
        balanced_category_path = os.path.join(balanced_dataset_path, category, subdir)
        os.makedirs(balanced_category_path, exist_ok=True)
        image_files = [os.path.join(category_path, f) for f in os.listdir(category_path) if f.endswith(('jpg', 'jpeg', 'png'))]
        selected_images = random.sample(image_files, min_count) if len(image_files) >= min_count else image_files
        for img_file in selected_images:
            img = cv2.imread(img_file)
            if img is None:
                continue
            img_save_path = os.path.join(balanced_category_path, os.path.basename(img_file))
            cv2.imwrite(img_save_path, img)
else:
    print("‚ö†Ô∏è No images found for balancing.")

# Display dataset size after balancing
balanced_counts, balanced_total = count_images_in_folders(balanced_dataset_path)
print(f"\nDataset Size After Balancing: {balanced_total} images")
visualize_distribution(balanced_counts, "Dataset Distribution After Balancing")
```

```python
# Splitting dataset into train, validation, and test
image_paths = []
labels = []
for category in categories.keys():
    category_path = os.path.join(balanced_dataset_path, category, categories[category])
    if os.path.isdir(category_path):
        for img_file in os.listdir(category_path):
            image_paths.append(os.path.join(category_path, img_file))
            labels.append(category)

# Train-Validation-Test Split
if image_paths:
    train_paths, test_paths, train_labels, test_labels = train_test_split(image_paths, labels, test_size=0.2, stratify=labels, random_state=42)
    train_paths, val_paths, train_labels, val_labels = train_test_split(train_paths, train_labels, test_size=0.2, stratify=train_labels, random_state=42)
    
    print(f"\nDataset Split:")
    print(f"   Training: {len(train_paths)} images")
    print(f"   Validation: {len(val_paths)} images")
    print(f"   Testing: {len(test_paths)} images")
    
    # Visualizing the dataset distribution after splitting
    split_counts = {"Training": len(train_paths), "Validation": len(val_paths), "Testing": len(test_paths)}
    plt.figure(figsize=(8, 5))
    sns.barplot(x=list(split_counts.keys()), y=list(split_counts.values()))
    plt.title("Dataset Split Distribution")
    plt.xlabel("Dataset Split")
    plt.ylabel("Number of Images")
    plt.show()
    
    # Visualizing class distribution in each split
    def plot_class_distribution(labels, title):
        class_counts = {label: labels.count(label) for label in set(labels)}
        plt.figure(figsize=(8, 5))
        sns.barplot(x=list(class_counts.keys()), y=list(class_counts.values()))
        plt.title(title)
        plt.xlabel("Category")
        plt.ylabel("Number of Images")
        plt.show()
    
    plot_class_distribution(train_labels, "Class Distribution in Training Set")
    plot_class_distribution(val_labels, "Class Distribution in Validation Set")
    plot_class_distribution(test_labels, "Class Distribution in Test Set")
    
else:
    print("‚ö†Ô∏è No images found for dataset splitting.")
```
